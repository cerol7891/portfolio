```python

df_group = df.groupby("area").mean().reset_index() 
xpos = np.arange(len(df_group))
plt.xticks(xpos, df_group["area"])
plt.bar(xpos,df_group["price"])
plt.title("Price Change By Area")
```

```python
df_age_group = df.groupby("age").mean("price")
plt.plot(df_age_group)
plt.xlabel("Age")
plt.ylabel("Price")
plt.title("Apartment Prices (Million TL)")
```

```python
df_floor = df.groupby("floor").mean("price").reset_index()
xpos_flor = np.arange(len(df_floor))
plt.bar(xpos_flor, df_floor["price"])
plt.xticks(xpos_flor, df_floor["floor"])
plt.xlabel("Level")
plt.ylabel("Price (TL)")
plt.title("Apartment Prices by Level")
plt.yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])
```

```python
df_neighborhood = df.groupby("neighborhood").mean("price").reset_index().sort_values("price", ascending=False).drop(index=15)
xpos_neighborhood = np.arange(len(df_neighborhood))
plt.xticks(xpos_neighborhood, df_neighborhood["neighborhood"],rotation=45 )
plt.bar(xpos_neighborhood,df_neighborhood["price"])
plt.xlabel("Neighborhood")
plt.ylabel("Price in TL")
plt.title("Average Apartmet Prices by Neighborhood")
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])
plt.rcParams["figure.figsize"] = (20,7)
```
