```python
#Diabetes vs Nondiabetes
sns.set_palette("colorblind")
plt.figure(figsize=(10,8))
plt.pie(df.groupby("Outcome").size(), autopct='%1.2f%%', labels=("Non-Diabetes","Diabetes"), colors=("Green", "Red"), radius=1, textprops={"size":"x-large", "fontweight":"bold"})
plt.title("Diabetes vs Non-Diabetes", fontsize=16, fontweight="bold")

plt.show()
```

```python
#Pregnancies and Glucose Level
plt.figure(figsize=(12,6))
plt.plot(df.groupby("Pregnancies")["Glucose"].mean())
plt.title("Glucose level by Number Of Pregnancies", fontsize=18, fontweight="bold" )
plt.xticks(np.arange(0,18))
plt.xlabel("Number of Pregnancy", fontsize=13, fontweight="bold")
plt.ylabel("Glucose Level", fontsize=13, fontweight="bold")
```

```python
#Glucose Level and Blood Preassure
plt.figure(figsize=(12,8))
scatter = plt.scatter(df["BloodPressure"], 
                df["Glucose"],
                alpha=0.8,
                c=df["Outcome"],
                cmap="viridis")
plt.legend(*scatter.legend_elements(),
                    loc="upper left",
                    title="Diabetes Status")
plt.xlabel("Blood Preassure", fontsize=13, fontweight="bold")
plt.ylabel("Glucose Level", fontsize=13, fontweight="bold")
plt.title("Blood Preassure - Glucose Level Relationship", fontsize=18, fontweight="bold")

plt.show()
```

```python
#Glucose Insulin Relationship
plt.figure(figsize=(12,8))
scatter = plt.scatter(df["Glucose"],
            df["Insulin"],
            alpha=0.8,
            c=df["Outcome"],
            cmap="copper"
             )
plt.legend(*scatter.legend_elements(),
                    loc="upper left",
                    title="Diabetes Status")
plt.title("Glucose Insulin Level Relationship", fontsize=18, fontweight="bold")
plt.xlabel("Glucose Level", fontsize=13, fontweight="bold")
plt.ylabel("Insulin Level", fontsize=13, fontweight="bold")
```

```python
#Diabetes Pedigree Glucose Level Relationship
plt.figure(figsize=(12,8))
scatter = plt.scatter(df["DiabetesPedigreeFunction"],
            df["Glucose"],
            alpha=0.8,
            c=df["Outcome"],
            cmap="PiYG")
plt.legend(*scatter.legend_elements(),
            loc="lower right",
            title="Diabetes Status")
plt.title("Diabetes Pedigree - Glucose Level Relationship", fontsize=18, fontweight="bold")
plt.xlabel("Diabetes Pedigree", fontsize=13, fontweight="bold")
plt.ylabel("Glucose Level", fontsize=13, fontweight="bold")
```

```python
#BMI & Glucose Level Relationship
plt.figure(figsize=(12,8))
scatter = plt.scatter(df["BMI"],
            df["Glucose"],
            alpha=0.8,
            c=df["Outcome"],
            cmap="PiYG")
plt.legend(*scatter.legend_elements(),
            loc="lower right",
            title="Diabetes Status")
plt.title("BMI - Glucose Level Relationship", fontsize=18, fontweight="bold")
plt.xlabel("BMI", fontsize=13, fontweight="bold")
plt.ylabel("Glucose Level", fontsize=13, fontweight="bold")
```

```python
#Age & Glucose Level Relationship
plt.figure(figsize=(12,8))
scatter = plt.scatter(df["Age"],
            df["Glucose"],
            alpha=0.8,
            c=df["Outcome"],
            cmap="PiYG")
plt.legend(*scatter.legend_elements(),
            loc="lower right",
            title="Diabetes Status")
plt.title("Age - Glucose Level Relationship", fontsize=18, fontweight="bold")
plt.xlabel("Age", fontsize=13, fontweight="bold")
plt.ylabel("Glucose Level", fontsize=13, fontweight="bold")
```
